```python
from tree_dataset import TreeDataset, show_img
import model as m
from torch.utils.data import DataLoader
import os
import torchvision.transforms as transforms
from d2lvit import *
import copy
import torch.nn as nn
import torch
import numpy as np
```

```python
train_set = TreeDataset(os.path.join('..', 'data', 'variety_2k'), m.resnet_preprocess()) 
val_set = TreeDataset(os.path.join('..', 'data', 'variety_1k'), m.resnet_preprocess())
print(f'Train size: {len(train_set)} Val size: {len(val_set)}')
train_loader = DataLoader(train_set, batch_size=32)
val_loader = DataLoader(val_set, batch_size=32)
device = m.get_device()
config = {'labels_key': 'tree_label'}
```

```python
d2l_digitsmodel = torch.load(os.path.join('..','models','d2lvit','digit-model.pt'))
d2l_treemodel = torch.load(os.path.join('..','models','d2lvit','tree-model.pt'))
torch_digitsmodel = torch.load(os.path.join('..','models','finetune','digit-model.pt'))
torch_treemodel = torch.load(os.path.join('..','models','finetune','tree-model.pt'))
```

```python
torch_train_acc = m.predict(torch_treemodel, train_loader, device, config, torch_digitsmodel, True)
print(torch_train_acc)
```

```python
torch_val_acc = m.predict(torch_treemodel, val_loader, device, config, torch_digitsmodel, True)
print(torch_val_acc)
```
