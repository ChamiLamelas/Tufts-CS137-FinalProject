```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
d2l_digit_train = np.load(os.path.join('..', 'models', 'd2lvit_3', 'digit_train_loss.npy'))
d2l_digit_val = np.load(os.path.join('..', 'models', 'd2lvit_3', 'digit_val_acc.npy'))
finetuned_digit_train = np.load(os.path.join('..', 'models', 'finetune_3', 'digit_train_loss.npy'))
finetuned_digit_val = np.load(os.path.join('..', 'models', 'finetune_3', 'digit_val_acc.npy'))
untrained_digit_train = np.load(os.path.join('..', 'models', 'untrained_torchvit', 'digit_train_loss.npy'))
untrained_digit_val = np.load(os.path.join('..', 'models', 'untrained_torchvit', 'digit_val_acc.npy'))
```

```python
d2l_tree_train = np.load(os.path.join('..', 'models', 'd2lvit_3', 'tree_train_loss.npy'))
d2l_tree_val = np.load(os.path.join('..', 'models', 'd2lvit_3', 'tree_val_acc.npy'))
finetuned_tree_train = np.load(os.path.join('..', 'models', 'finetune_3', 'tree_train_loss.npy'))
finetuned_tree_val = np.load(os.path.join('..', 'models', 'finetune_3', 'tree_val_acc.npy'))
untrained_tree_train = np.load(os.path.join('..', 'models', 'untrained_torchvit', 'tree_train_loss.npy'))
untrained_tree_val = np.load(os.path.join('..', 'models', 'untrained_torchvit', 'tree_val_acc.npy'))

d2l_tree_noprims_train = np.load(os.path.join('..', 'models', 'd2lvit_noprims', 'tree_train_loss.npy'))
d2l_tree_noprims_val = np.load(os.path.join('..', 'models', 'd2lvit_noprims', 'tree_val_acc.npy'))
finetuned_tree_noprims_train = np.load(os.path.join('..', 'models', 'finetune_noprims', 'tree_train_loss.npy'))
finetuned_tree_noprims_val = np.load(os.path.join('..', 'models', 'finetune_noprims', 'tree_val_acc.npy'))
untrained_tree_noprims_train = np.load(os.path.join('..', 'models', 'untrained_torchvit_noprims', 'tree_train_loss.npy'))
untrained_tree_noprims_val = np.load(os.path.join('..', 'models', 'untrained_torchvit_noprims', 'tree_val_acc.npy'))
```

```python
epochs = np.arange(1, 101)
```

```python
plt.figure()
plt.plot(epochs, d2l_digit_train, label='D2L ViT')
plt.plot(epochs, finetuned_digit_train, 'g--', label='Fine Tuned PyTorch ViT')
plt.plot(epochs, untrained_digit_train, 'r-.', label='Scratch Trained PyTorch ViT')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid()
plt.legend()
plt.title('Training Loss on extra_variety_4k Digits')
plt.savefig(os.path.join('..', 'plots', 'digit_train_loss.png'))
plt.show()
```

```python
start = 20
later = epochs[start:]
plt.figure()
plt.plot(later, d2l_digit_train[start:], label='D2L ViT')
plt.plot(later, finetuned_digit_train[start:], 'g--', label='Fine Tuned PyTorch ViT')
plt.plot(later, untrained_digit_train[start:], 'r-.', label='Scratch Trained PyTorch ViT')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid()
plt.legend()
plt.title('Training Loss on extra_variety_4k Digits Closer Look')
plt.savefig(os.path.join('..', 'plots', 'digit_train_loss_later.png'))
plt.show()
```

```python
plt.figure()
plt.plot(epochs, d2l_digit_val, label='D2L ViT')
plt.plot(epochs, finetuned_digit_val, 'g--', label='Fine Tuned PyTorch ViT')
plt.plot(epochs, untrained_digit_val, 'r-.', label='Scratch Trained PyTorch ViT')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid()
plt.legend()
plt.title('Validation Accuracy on extra_variety_2k Digits')
plt.savefig(os.path.join('..', 'plots', 'digit_val_acc.png'))
plt.show()
```
